```python
import pandas as pd
import numpy as np

# Load the data
data = pd.read_excel('rolling_stones_spotify.xlsx')

# Check for duplicates
duplicates = data[data.duplicated()]
print('Number of duplicates:', len(duplicates))

# Check for missing values
missing_values = data.isnull().sum()
print('Number of missing values in each column:', missing_values)

# Check for irrelevant (erroneous entries) values
irrelevant_values = data[data['popularity'] < 0]
print('Number of songs with negative popularity:', len(irrelevant_values))

# Check for outliers
outliers = data[(data['acousticness'] > 1) | (data['danceability'] > 1) | (data['energy'] > 1) | (data['instrumentalness'] > 1) | (data['liveness'] > 1) | (data['speechiness'] > 1)]
print('Number of songs with outlier feature values:', len(outliers))
```

```python
# Remove duplicates
data = data.drop_duplicates()

# Remove irrelevant values
data = data.drop(index=irrelevant_values.index)

# Remove outliers
data = data.drop(index=outliers.index)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Find out which two albums should be recommended to anyone based on the number of popular songs in an album
sns.barplot(x='album', y='popularity', data=data.sort_values(by='popularity', ascending=False))
plt.xlabel('Album')
plt.ylabel('Popularity')
plt.title('Number of popular songs in each album')
plt.show()

# Explore the relationship between popularity and other features
sns.heatmap(data.corr(), annot=True)
plt.title('Correlation matrix of song features')
plt.show()

# Explore how popularity has changed over time
sns.lineplot(x='release_date', y='popularity', data=data)
plt.xlabel('Release date')
plt.ylabel('Popularity')
plt.title('Popularity of songs over time')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data[['danceability', 'energy']])

# Identify the right number of clusters
cluster_sizes = range(2, 11)
silhouette_scores = []
for cluster_size in cluster_sizes:
    kmeans = KMeans(n_clusters=cluster_size)
    labels = kmeans.fit_predict(data_pca)
    silhouette_score = silhouette_score(data_pca, labels)
    silhouette_scores.append(silhouette_score)

plt.plot(cluster_sizes, silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Elbow method to select the number of clusters')
plt.show()

# Train the KMeans model
kmeans = KMeans(n_clusters=6)
labels = kmeans.fit_predict(data_pca)

# Define each cluster based on the features
cluster_means = data_pca.groupby('cohort').mean()
print(cluster_means)
```
